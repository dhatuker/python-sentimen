# Impor Library

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from nltk.corpus import stopwords 
import string
import re
import numpy as np
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

# Insert Data Training

```python
#file csv untuk data training dan testing sudah dibedakan, 160 data untuk training dan 20 data untuk testing.
#Untuk testing 20 komentar positif, 20 komentar negatif
#read data
data_training = pd.read_csv('D:/data_training.csv',sep=";",encoding = 'unicode_escape')
data_training.columns = [
    "Komentar", "Sumber", "Pakar 1",
    "Pakar 2", "Pakar 3", "Pakar 4",
    "Positif", "Negatif", "Hasil Akhir"]
## di punyaku tuh encoding nya error, jadi aku tambahin encoding = 'unicode_escape'
data_training.head(160)
```

```python
#ambil matriks kolom Komentar
komentar=np.array(data_training['Komentar'])
print("Data Training Komentar : \n",komentar)
```

# Preprocessing

```python
#Preprocessing
factory = StemmerFactory()
stemmer = factory.create_stemmer()
listStopword =  set(stopwords.words('indonesian'))
case_fold = []
for i in komentar:
    if i not in listStopword:
        lowercase = i.translate(str.maketrans("","",string.punctuation)).lower()
        hapus_angka = re.sub(r"\d+", "", lowercase)
        hapus_space = hapus_angka.strip()
        stemming = stemmer.stem(hapus_space)
        case_fold.append(stemming)
print("Hasil Preprocessing : \n",case_fold)
```

# Pembobotan

```python
#Pembobotan
vectorizer = TfidfVectorizer(stop_words=listStopword)
x_train = vectorizer.fit_transform(case_fold)
features = vectorizer.get_feature_names()
dense = x_train.todense()
listfeatures = dense.tolist()

df = pd.DataFrame(listfeatures, columns=features)
print("Hasil Pembobotan : ")
df.head(160)
```

# Klasifikasi

```python
# Label untuk data training
y_train = np.array(data_training['Hasil Akhir'])
# Clustering the document with KNN classifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)
```

```python
#testing
data_testing = pd.read_csv('D:/data_testing.csv',sep=";",encoding = 'unicode_escape')
data_testing.columns = [
    "Komentar", "Sumber", "Pakar 1",
    "Pakar 2", "Pakar 3", "Pakar 4",
    "Positif", "Negatif", "Hasil Akhir"]
testing_kalimat = data_testing['Komentar']
bersihkan_kalimat = []
for test in testing_kalimat:
    if test not in listStopword:
        lowercase_test = test.translate(str.maketrans("","",string.punctuation)).lower()
        hapus_angka_test = re.sub(r"\d+", "", lowercase_test)
        hapus_space_test = hapus_angka_test.strip()
        stemming_test = stemmer.stem(hapus_space_test)
        bersihkan_kalimat.append(stemming_test)
print("Preprocessing Komentar Test : \n",bersihkan_kalimat)
```

```python
x_test = vectorizer.transform(bersihkan_kalimat) 
y_pred = classifier.predict(x_test)
y_test = np.array(data_testing['Hasil Akhir'])
```

```python
TP = 0
TN = 0
FN = 0
FP = 0
print ("\n-------------------------------PREDICTIONS BY KNN------------------------------------------")
for i in range(len(testing_kalimat)):
    print(i+1,". Komentar : ",testing_kalimat[i])
    if(y_test[i] == 'Positif'):
        if(y_pred[i] == 'Positif'):
            TP +=1
        elif(y_pred[i] == 'Negatif'):
            FP +=1
    else:
        if(y_pred[i] == 'Negatif'):
            TN +=1
        elif(y_pred[i] == 'Positif'):
            FN += 1
    print("\tHasil Prediksi\t: ",y_pred[i])
    print("\tSebenarnya\t: ",y_test[i])
    print("==========================================================================================")
print("TP : ",TP,"\nTN : ",TN,"\nFN : ",FN,"\nFP : ",FP)
```

# Menghitung Akurasi

```python
#hitung akurasi

from sklearn.metrics import classification_report, confusion_matrix 
print(confusion_matrix(y_test, y_pred)) 
print(classification_report(y_test, y_pred))
```
